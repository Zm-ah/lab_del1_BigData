```sql
-- Verification 1: Country with most finishers in km races
SELECT a.athlete_country, COUNT(*) AS total_finishers
FROM marathos.gold.fct_results f
JOIN marathos.gold.dim_athlete a ON f.athlete_key = a.athlete_key
JOIN marathos.gold.dim_event e ON f.event_id = e.event_id
WHERE e.event_type = 'kilometers'
GROUP BY a.athlete_country
ORDER BY total_finishers DESC
LIMIT 1
```

```sql
-- Verification 2: Average speed for female athletes
SELECT ROUND(AVG(f.athlete_average_speed), 2) AS avg_speed_female
FROM marathos.gold.fct_results f
JOIN marathos.gold.dim_athlete a ON f.athlete_key = a.athlete_key
WHERE a.athlete_gender = 'F'
```

```sql
-- Verification 3: Unique athletes in 2019
SELECT COUNT(DISTINCT f.athlete_key) AS unique_athletes_2019
FROM marathos.gold.fct_results f
JOIN marathos.gold.dim_time t ON f.year_id = t.year_id
WHERE t.year_of_event = 2019
```

```sql
-- Verification 4: Top 5 fastest athletes
SELECT athlete_id, athlete_country, athlete_gender,
       ROUND(athlete_average_speed, 2) AS avg_speed_kmh
FROM marathos.gold.mart_fastest_athletes_km
ORDER BY athlete_average_speed DESC
LIMIT 5


-- FINDING: Genie used MAX(speed) while our mart table uses average speed.
-- This explains the different rankings. Our mart_fastest_athletes_km 
-- is based on athlete_average_speed, not peak speed.
-- Genie result: SWE athlete 394189 (28.224 km/h max)
-- SQL result: NZL athlete 868574 (24.79 km/h avg)
-- Conclusion: Results differ due to different speed metrics used.
```
