```sql
%sql
-- Fråga: "Which European country has the most ultramarathon finishers?"
SELECT athlete_country, total_finishers
FROM marathos.gold.mart_top_countries
ORDER BY total_finishers DESC
LIMIT 1
```

```sql
%sql
-- Fråga: "How has participation in European ultramarathons changed over the years?"
SELECT year_of_event, COUNT(*) AS total_results
FROM marathos.gold.fct_results f
JOIN marathos.gold.dim_time t ON f.year_id = t.year_id
GROUP BY year_of_event
ORDER BY year_of_event ASC
```

```sql
%sql
-- Fråga: "What is the gender distribution among European ultramarathon athletes?"
SELECT athlete_gender, COUNT(*) AS number_of_results
FROM marathos.gold.fct_results f
JOIN marathos.gold.dim_athlete a ON f.athlete_key = a.athlete_key
WHERE athlete_gender IS NOT NULL
GROUP BY athlete_gender
```

```sql
%sql
-- Fråga: "Who are the fastest athletes in European ultramarathons?"
SELECT athlete_country, athlete_gender, event_name, athlete_average_speed
FROM marathos.gold.mart_fastest_athletes
ORDER BY athlete_average_speed DESC
LIMIT 10
```

```sql
%sql
-- Fråga: "What is the average speed by event distance?"
SELECT event_distance_length, ROUND(AVG(athlete_average_speed), 2) AS avg_speed_kmh
FROM marathos.gold.fct_results f
JOIN marathos.gold.dim_event e ON f.event_id = e.event_id
GROUP BY event_distance_length
ORDER BY avg_speed_kmh DESC
```

```sql
%sql
-- Fråga: "Which countries have the highest average finishing speed?"
SELECT athlete_country, avg_speed_kmh
FROM marathos.gold.mart_top_countries
ORDER BY avg_speed_kmh DESC
LIMIT 10
```
