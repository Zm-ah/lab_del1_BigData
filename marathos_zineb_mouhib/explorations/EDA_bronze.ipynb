- How are ages distributed among the runners?

```python
VOLUME_PATH = "/Volumes/marathos/bronze/raw/TWO_CENTURIES_OF_UM_RACES.csv"
```

```python

df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(VOLUME_PATH)
    
)

display(df)
```

```python
# Find out number of rows
print(f"Rows: {df.count()}")
```

```python
# find out number of columns 
print(f"Columns: {len(df.columns)}")
```

```python
# print the shcema of the dataframe
df.printSchema()
```

```python
#Count nulls
from pyspark.sql.functions import col, sum

null_counts = df.select([
    sum(col(c).isNull().cast("int")).alias(c)
    for c in df.columns
])

display(null_counts)
```

```python
# Describe numerical columns
display(df.describe())
```

```python
# unic events 
df.select("Event name").distinct().count()

# Unique events
# The dataset contains 26,907 unique marathon and ultramarathon events based on the column Event name.
```

```python
# countries most represented in the dataset
display(
    df.groupBy("Athlete Country")
    .count()
    .orderBy(col("count"), ascending=False)
)
```

```python
# age distribution
from pyspark.sql.functions import col, year, current_date

df_age = df.withColumn(
    "age", year(current_date()) - col("Athlete year of birth")
)
display(
    df_age.groupBy("age")
    .count()
    .orderBy(col("age"))
)
```

```python
display(df.select("Athlete year of birth").distinct().orderBy("Athlete year of birth"))
```
